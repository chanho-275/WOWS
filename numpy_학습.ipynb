```python
#gemini ì˜ˆì œ 5
#ì£¼ê°€ ê²½ë¡œ ì‹œë®¬ë ˆì´ì…˜ì˜ ê¸°ì´ˆ

# 5ì¼ ë™ì•ˆ ì£¼ê°€ ê²½ë¡œë¥¼ ì €ì¥í•  ë°°ì—´ ì´ˆê¸°í™” (ì‹œì‘ ì£¼ê°€ 100)
# np.zeros((í–‰, ì—´))ì„ ì‚¬ìš©í•˜ì—¬ ë¹ˆ ê³µê°„ ìƒì„±
stock_path = np.zeros(5)
stock_path[0] = 100 # ì²«ë‚  ì£¼ê°€ ì„¤ì •

daily_shock = np.array([1.5, -0.8, 2.1, -0.3]) # ì„ì˜ì˜ ì¼ì¼ ì¶©ê²©

# for ë¬¸ì„ ì‚¬ìš©í•˜ì—¬ ì£¼ê°€ ë³€í™” ê³„ì‚°
for i in range(1, 5):
    # ì „ë‚  ì£¼ê°€ì— ì¼ì¼ ì¶©ê²©ë§Œí¼ ë”í•¨ (+= í™œìš©)
    stock_path[i] = stock_path[i-1] + daily_shock[i-1]

print("5ì¼ê°„ì˜ ì£¼ê°€ ê²½ë¡œ:", stock_path)
# ì¶œë ¥: 5ì¼ê°„ì˜ ì£¼ê°€ ê²½ë¡œ: [100.  101.5  100.7  102.8  102.5]

# ------------------------------------------------------
#ì‘ìš© ì—°ìŠµ 5-1 :ì£¼ê°€ ê²½ë¡œ, ì¡°ì • ì£¼ê°€ ê²½ë¡œ êµ¬í•˜ê¸°
stock=np.zeros(5)
ad_stock=np.zeros(5)
stock[0]=100000
ad_stock[0]=100000
shock=np.array([0.076,-0.034,-0.027,0.011,-0.092])
div=np.array([0,0,0.08,0,0])
# ë°°ë‹¹ê¸ˆìœ¼ë¡œ ì¸í•œ ìˆ˜ìµë¥ , ì¶”í›„ì— ë³´ë‹¤ ì •í™•í•œ ê³µì‹ìœ¼ë¡œ ì¶”ê°€ ì½”ë”©í•´ë³´ê¸°

for i in range (1,5):
  stock[i]=stock[i-1]*(1+shock[i-1])
  ad_stock[i]=ad_stock[i-1]*(1+shock[i-1])+ad_stock[i-1]*(div[i-1])


print(f"5ì¼ê°„ì˜ ì£¼ê°€ ê²½ë¡œ :{stock}")
print(f"5ì¼ê°„ì˜ ì¡°ì • ì£¼ê°€ ê²½ë¡œ :{ad_stock}")
#print(f"5ì¼ê°„ì˜ ì¡°ì • ì£¼ê°€ ê²½ë¡œ:{ad_stock:.2f}") - ì—ëŸ¬
#np.arrayë¥¼ í†µí•´ ë§Œë“¤ì€ ë°°ì—´ì€ ndarray, ìì²´ì ìœ¼ë¡œ ë°ì´í„° íƒ€ì…ì„ ê°€ì§€ê³  ìˆìœ¼ë©° float ê°ì²´ì™€ëŠ” ë‹¤ë¥¸ í‘œê¸°ë²•
#.2f í¬ë§·íŒ…ì€ ë‹¨ì¼ ìˆ«ìë¥¼ ìœ„í•œ f-string ê¸°ëŠ¥ì´ì–´ì„œ ndarrayì— ì ìš©í•  ìˆ˜ ì—†ìŒ
```

```python
#gemini ì˜ˆì œ 6
#í¬íŠ¸í´ë¦¬ì˜¤ ìì‚° ì¤‘ë³µ ì œê±°

# íˆ¬ì í¬íŠ¸í´ë¦¬ì˜¤ Aì™€ Bì— í¬í•¨ëœ ì¢…ëª© ì½”ë“œ
portfolio_A = {"AAPL", "MSFT", "GOOG"}
portfolio_B = {"MSFT", "TSLA", "NVDA"}

# ë‘ í¬íŠ¸í´ë¦¬ì˜¤ë¥¼ í•©ì¹œ ì „ì²´ ê³ ìœ  ìì‚° ëª©ë¡ (í•©ì§‘í•©)
unique_assets = portfolio_A.union(portfolio_B)
print("ì „ì²´ ê³ ìœ  ìì‚°:", unique_assets)
# ì¶œë ¥: ì „ì²´ ê³ ìœ  ìì‚°: {'TSLA', 'GOOG', 'AAPL', 'MSFT', 'NVDA'}

# ë‘ í¬íŠ¸í´ë¦¬ì˜¤ì— ê³µí†µìœ¼ë¡œ í¬í•¨ëœ ìì‚° (êµì§‘í•©)
common_assets = portfolio_A.intersection(portfolio_B)
print("ê³µí†µ ìì‚°:", common_assets)
# ì¶œë ¥: ê³µí†µ ìì‚°: {'MSFT'}
```

```python
#gemini ì˜ˆì œ 7
#ì‹œë®¬ë ˆì´ì…˜ ì¤‘ë‹¨ ì¡°ê±´

# ë¦¬ìŠ¤í¬ ì§€í‘œê°€ ì„ê³„ì¹˜ë¥¼ ì´ˆê³¼í•  ë•Œ ì‹œë®¬ë ˆì´ì…˜ ì¤‘ë‹¨
risk_threshold = 5.0
current_risk = 3.0
day = 0

while current_risk < 10.0: # ìµœëŒ€ 10ì¼ ì‹œë®¬ë ˆì´ì…˜ ê°€ì •
    day += 1
    current_risk += np.random.rand() * 1.5 # 0 ~ 1.5 ì‚¬ì´ì˜ ëœë¤ ì¦ê°€

    if current_risk >= risk_threshold:
        print(f"ğŸš¨ {day}ì¼ì°¨ì— ë¦¬ìŠ¤í¬ ì§€í‘œ({current_risk:.2f})ê°€ ì„ê³„ì¹˜({risk_threshold:.1f})ë¥¼ ì´ˆê³¼í•˜ì—¬ ì¤‘ë‹¨í•©ë‹ˆë‹¤.")
        break # ë°˜ë³µë¬¸ ì¦‰ì‹œ ì¢…ë£Œ
# ì¶œë ¥ ì˜ˆì‹œ: ğŸš¨ 3ì¼ì°¨ì— ë¦¬ìŠ¤í¬ ì§€í‘œ(5.17)ê°€ ì„ê³„ì¹˜(5.0)ë¥¼ ì´ˆê³¼í•˜ì—¬ ì¤‘ë‹¨í•©ë‹ˆë‹¤. (ì‹¤í–‰í•  ë•Œë§ˆë‹¤ ë‹¤ë¦„)

# ------------------------------------------------------
#ì‘ìš© ì—°ìŠµ 7-1 : ì£¼ì‹ ë§¤ìˆ˜, ë§¤ë„ ì¡°ê±´ ì„¤ì •í•˜ê¸°

buy=8800
sell= 12000
stock=10000
day=0

while day <7 :
  day +=1
  stock += stock*(np.random.normal(0,0.07))
  #geminiê°€ ë³€ë™ì„± 7%ëŠ” ë¹„í˜„ì‹¤ì ì´ë¯€ë¡œ ë¡œê·¸ìˆ˜ìµë¥  ë˜ëŠ” ë‹¨ìˆœìˆ˜ìµë¥ ì„ ëŒ€ì…í•˜ëŠ” ê²ƒì´ ë°”ëŒì§í•˜ë‹¤ê³  ì¶©ê³ , ì—¬ê¸°ì„œëŠ” ìˆ˜ìµë¥ ì„ ì œëŒ€ë¡œ êµ¬í•˜ëŠ”ê²Œ ì¤‘ìš”í•œê±´ ì•„ë‹ˆë‹ˆê¹Œ pass

  if stock >= sell or stock <= buy :
    print(f"{day}ì¼ì°¨ ê±°ë˜ ë°œìƒ")

    if stock >= sell :
      print(f"ë§¤ë„ ê±°ë˜ ë°œìƒ, ë§¤ë„ê°€ : {stock}ì›")
      break
    else :
      print(f"ë§¤ìˆ˜ ê±°ë˜ ë°œìƒ, ë§¤ìˆ˜ê°€ : {stock}ì›")
      break
if day== 7 :
  print("7ì¼ê°„ ê±°ë˜ ë¯¸ì²´ê²°")
```

```python
#gemini ì˜ˆì œ 8
#íŠ¹ì • ì¡°ê±´ ì¶©ì¡± ì‹œ ê±´ë„ˆë›°ê¸°

# ì¼ë³„ ìˆ˜ìµë¥  ë¦¬ìŠ¤íŠ¸
returns = [0.005, -0.01, 0.02, -0.05, 0.001]
trade_days = 0

# ì†ì‹¤(-0.02 ì´í•˜)ì´ ë°œìƒí•œ ë‚ ì€ ì§‘ê³„ì—ì„œ ì œì™¸
for r in returns:
    if r <= -0.02:
        print(f"ì†ì‹¤({r:.3f}) ë°œìƒ. ê±°ë˜ì¼ ì§‘ê³„ì—ì„œ ì œì™¸.")
        continue # ì•„ë˜ ì½”ë“œ ì‹¤í–‰í•˜ì§€ ì•Šê³  ë‹¤ìŒ ë°˜ë³µìœ¼ë¡œ ë„˜ì–´ê°
    trade_days += 1

print(f"ìˆœìˆ˜ìµì´ ë°œìƒí•˜ê±°ë‚˜ ê²½ë¯¸í•œ ì†ì‹¤ì´ ë°œìƒí•œ ê±°ë˜ì¼ ìˆ˜: {trade_days}ì¼")
# ì¶œë ¥:
# ì†ì‹¤(-0.050) ë°œìƒ. ê±°ë˜ì¼ ì§‘ê³„ì—ì„œ ì œì™¸.
# ìˆœìˆ˜ìµì´ ë°œìƒí•˜ê±°ë‚˜ ê²½ë¯¸í•œ ì†ì‹¤ì´ ë°œìƒí•œ ê±°ë˜ì¼ ìˆ˜: 4ì¼

# ------------------------------------------------------
#ì‘ìš© ì—°ìŠµ 8-1 :ì¡°ê±´ì„ ë§Œì¡±í•˜ëŠ” ë³µìˆ˜ì˜ case ê°ì§€í•˜ê¸°
vol_lb=0.01
vol_ub=0.05
day=0
flag=False

while day <10 :
  day +=1
  vol=max(np.random.normal(0.03,0.01),0)
  if vol_lb < vol < vol_ub :
   continue
  flag=True
  if vol>= vol_ub :
    print(f"{day}ì¼ì°¨ ê³ ë³€ë™ì„± ì´ìƒê°’ ê°ì§€, ë³€ë™ì„± : {vol}")
  else :
    print(f"{day}ì¼ì°¨ ì €ë³€ë™ì„± ì´ìƒê°’ ê°ì§€, ë³€ë™ì„± : {vol}")

if flag == False :
  print(f"{day}ì¼ê°„ ë³€ë™ì„± ì´ìƒê°’ ë¯¸ê°ì§€")
```
